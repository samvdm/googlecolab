# Setting up Kaggle API

This note book describes the process of connecting to the Kaggle API in order to download datasets. Before you start ensure that you are first registered on [kaggle](https://www.kaggle.com). 

## Get kaggle.json
Once your account is created follow the following steps:



1.   Click on your profile icon (top left corner) and select "myprofie)
2.   On the menu click on more shown as "..." and then select account
3. Scroll to the middle until you get to a section with the header API and select "“Create a New API Token”

The last action will automatically download a kaggle.json file to your local drive.
